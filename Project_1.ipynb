```python
active = newdata.loc[newdata['new_value'] <= 10000]
active
```

```python
#For the inactive molecule
inactive = newdata.loc[newdata['new_value'] >= 10000]
inactive=inactive.sort_values("new_value", ascending=True).reset_index(drop=True)
inactive
```
